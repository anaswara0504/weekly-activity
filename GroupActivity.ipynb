### 3. Examine the variables ‘Hardness’, ‘Sulfate’, and ‘Organic_carbon’ and comment on your observations.

```python
import seaborn as sns
import matplotlib.pyplot as plt
d1 = d[['Hardness','Sulfate','Organic_carbon']]
sns.pairplot(d1)
```

```python
d1.describe()
```

```python
cm = d1.corr()
plt.subplots(figsize = (6,4))
sns.heatmap(cm,vmin=-1,vmax=1,annot=True,cmap='Blues')
```

### Inferences
* Hardness, Sulfate, and Organic_carbon are normal distributed.
* Mean and standard deviation of sulfate is higher than hardness.
* Mean and standard deviation of organic carbon is very much lower than Hardness and Sulfate.
* Hardness and Sulfate have low negative correlation with each other.
* Hardness and Organic carbon have very low positive correlation with each other.
* Sulfate and Organic carbon have very low positive correlation with each other.

### 4. Carry out correlation analysis on the dataset and find the correlation factor between
### a. ph and Chloramines
### b. ph and Conductivity
### What do we infer from this correlation value

```python
d2 = d[['Chloramines','ph']]
cm = d2.corr()
plt.subplots(figsize = (6,4))
sns.heatmap(cm,vmin=-1,vmax=1,annot=True,cmap='Greens')

d2 = d[['Conductivity','ph']]
cm = d2.corr()
plt.subplots(figsize = (6,4))
sns.heatmap(cm,vmin=-1,vmax=1,annot=True,cmap='Reds')
```

### Inference
* Chloramies have very low negative correlation with ph.
* Conductivity have very low positive correlation with ph.
